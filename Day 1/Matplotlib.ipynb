# Scatter plot

```python
a=np.random.random(10)
b=np.random.random(10)
```

```python
a
```

```python
b
```

```python
plt.figure(figsize=(10,3))
plt.scatter(a,b,c="r",marker="X",alpha=0.9)
plt.xlabel("This is X-Axis")
plt.ylabel("This is Y-Axis")
plt.title("X Vs Y")
```

# Line Graph

```python
a=np.linspace(1,10,15)
```

```python
a
```

```python
b=np.sin(a)
```

```python
b
```

```python
plt.figure(figsize=(10,3))
plt.plot(a,b,"--r",alpha=0.9)
plt.xlabel("This is X-Axis")
plt.ylabel("This is Y-Axis")
plt.title("X Vs Y")
```

# BarGraph

```python
x=["Dhoni","kohli","Rohit","Rahul","SKY"]
```

```python
x
```

```python
runs=[34,56,48,21,32]
```

```python
runs
```

```python
plt.figure(figsize=(10,3))
plt.bar(x,runs,color="r")
plt.xlabel("This is X-Axis")
plt.ylabel("This is Y-Axis")
plt.title("X Vs Y")
```

# BarGraph Horizontally

```python
plt.figure(figsize=(10,3))
plt.barh(x,runs,color="r")
plt.xlabel("This is X-Axis")
plt.ylabel("This is Y-Axis")
plt.title("X Vs Y")
```

# 3D Graphs

```python
x=np.random.random(50)
y=np.random.random(50)
z=np.random.random(50)
```

```python
x,y,z
```

```python
fig=plt.figure()
ax=fig.add_subplot(projection="3d")
ax.scatter(x,y,z,alpha=1.0,c="r")
plt.xlabel("This is X-Axis")
plt.ylabel("This is Y-Axis")
plt.title("X Vs Y")
plt.show()
```
